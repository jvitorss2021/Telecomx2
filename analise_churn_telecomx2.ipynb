# 1. Preparação dos Dados

```python
# CÓDIGO INICIAL DO SEU PROJETO 2

import pandas as pd

# Carrega a base de dados tratada que você fez o upload
df = pd.read_csv('telecom_dados_tratados_final.csv')

# Verifica se os dados foram carregados corretamente
print("Base de dados tratada do Projeto 1 carregada com sucesso!")
df.head()
```

```python
# Drop columns that are not useful for analysis or predictive models
df = df.drop('customerID', axis=1, errors='ignore')

# Verify the columns remaining in the DataFrame
print("Columns after dropping 'customerID':")
print(df.columns)
```

```python
# PASSO 1: Identificar as colunas categóricas que ainda são texto (object)
colunas_categoricas = df.select_dtypes(include=['object']).columns

# PASSO 2: Aplicar o One-Hot Encoding usando a melhor prática
df_encoded = pd.get_dummies(df, columns=colunas_categoricas, drop_first=True)

# PASSO 3: Verificar o resultado
print("DataFrame após o One-Hot Encoding (com drop_first=True):")
display(df_encoded.head())

print("\nDimensões do novo DataFrame:", df_encoded.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------------------------------------------------------
# PASSO 1: Calcular a contagem e a proporção das classes
# ---------------------------------------------------------------------------
# Usamos value_counts() para contar as ocorrências de cada classe (0 e 1)
contagem_churn = df_encoded['Churn'].value_counts()
proporcao_churn = df_encoded['Churn'].value_counts(normalize=True) * 100

print("--- Análise do Desequilíbrio de Classes ---")
print("\nContagem de Clientes por Classe:")
# O .get(0, 0) garante que não dará erro se uma classe não existir
print(f"Permaneceram (Classe 0): {contagem_churn.get(0, 0)}")
print(f"Evadiram (Classe 1):     {contagem_churn.get(1, 0)}")
print("\nProporção de Clientes por Classe:")
print(f"Permaneceram (Classe 0): {proporcao_churn.get(0, 0):.2f}%")
print(f"Evadiram (Classe 1):     {proporcao_churn.get(1, 0):.2f}%")
print("\n" + "="*80 + "\n")


# ---------------------------------------------------------------------------
# PASSO 2: Visualizar o desequilíbrio
# ---------------------------------------------------------------------------
# Um gráfico de barras é a melhor forma de visualizar essa diferença.
print("--- Gráfico de Distribuição das Classes ---")
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=df_encoded, palette=['#55A868', '#C44E52'])
plt.title('Distribuição de Clientes (Churn vs. Permaneceu)', fontsize=16)
plt.ylabel('Número de Clientes', fontsize=12)
plt.xlabel('Status do Cliente', fontsize=12)
plt.xticks([0, 1], ['Permaneceram (0)', 'Evadiram (1)']) # Renomeia os eixos para clareza
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# ---------------------------------------------------------------------------
# PASSO 1: Separar as variáveis preditivas (X) da variável-alvo (y)
# ---------------------------------------------------------------------------
# X contém todas as colunas, EXCETO 'Churn'
X = df_encoded.drop('Churn', axis=1)

# y contém APENAS a coluna 'Churn'
y = df_encoded['Churn']

print("--- Dados separados em Features (X) e Alvo (y) ---")
print("Dimensões de X:", X.shape)
print("Dimensões de y:", y.shape)
print("\n" + "="*80 + "\n")


# ---------------------------------------------------------------------------
# PASSO 2: Aplicar a Padronização (Standardization)
# ---------------------------------------------------------------------------
# 1. Cria o objeto scaler
scaler = StandardScaler()

# 2. Treina o scaler nos dados e transforma X
X_scaled = scaler.fit_transform(X)

# 3. O resultado é um array NumPy. Vamos convertê-lo de volta para um DataFrame
#    para facilitar a visualização, mantendo os nomes das colunas.
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)


# ---------------------------------------------------------------------------
# PASSO 3: Verificar o resultado
# ---------------------------------------------------------------------------
print("--- Verificação do DataFrame Após a Padronização ---")
print("Primeiras 5 linhas do DataFrame padronizado (X_scaled_df):")
display(X_scaled_df.head())

print("\nEstatísticas descritivas para confirmar a padronização:")
# Observe como a média (mean) de todas as colunas está próxima de 0
# e o desvio padrão (std) está próximo de 1.
display(X_scaled_df.describe())
```

# 2. Análise Exploratória e Seleção de Variáveis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# ---------------------------------------------------------------------------
# PASSO 1: Juntar X e y para a análise de correlação
# ---------------------------------------------------------------------------
# O método .corr() funciona melhor em um único DataFrame.
# Como os índices foram preservados, podemos concatená-los lado a lado.
df_para_correlacao = pd.concat([X_scaled_df, y], axis=1)


# ---------------------------------------------------------------------------
# PASSO 2: Calcular a matriz de correlação
# ---------------------------------------------------------------------------
correlation_matrix = df_para_correlacao.corr()


# ---------------------------------------------------------------------------
# PASSO 3: Focar nas correlações com a variável-alvo "Churn"
# ---------------------------------------------------------------------------
# Uma matriz de correlação completa pode ser muito grande e difícil de ler.
# O mais importante é ver a correlação de cada variável com 'Churn'.
# Vamos extrair essa informação e ordenar para ver as mais importantes.

corr_churn = correlation_matrix['Churn'].sort_values(ascending=False)

print("--- Correlação das Variáveis com o Churn ---")
print("(Valores positivos indicam que a variável aumenta a chance de Churn)")
print("(Valores negativos indicam que a variável diminui a chance de Churn)")
print("\n" + "="*80 + "\n")
print(corr_churn)


# ---------------------------------------------------------------------------
# PASSO 4: Visualizar as correlações mais fortes em um gráfico
# ---------------------------------------------------------------------------
# Vamos plotar apenas as 10 mais fortes positivamente e as 10 mais fortes negativamente
# para um gráfico mais limpo e informativo.

# Selecionando as top 10 positivas (excluindo o próprio 'Churn', que é 1.0)
top_10_pos = corr_churn[1:11]

# Selecionando as top 10 negativas
top_10_neg = corr_churn[-10:]

# Concatenando as duas séries para plotagem
top_corr = pd.concat([top_10_pos, top_10_neg])

plt.figure(figsize=(10, 8))
sns.barplot(x=top_corr.values, y=top_corr.index, palette="coolwarm")
plt.title('Top 20 Variáveis Mais Correlacionadas com o Churn', fontsize=16)
plt.xlabel('Coeficiente de Correlação', fontsize=12)
plt.ylabel('Variáveis', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

# 3. Modelagem Preditiva e Avaliação

```python
from sklearn.model_selection import train_test_split

# ---------------------------------------------------------------------------
# PASSO 1: Realizar a divisão treino-teste
# ---------------------------------------------------------------------------
# Nossos inputs são:
# X_scaled_df: O DataFrame com todas as features (variáveis preditivas) já tratadas e padronizadas.
# y: A Series com a nossa variável-alvo (Churn).

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df,
    y,
    test_size=0.3,       # 30% dos dados para o conjunto de teste
    random_state=42,     # Semente para garantir a mesma divisão sempre
    stratify=y           # Parâmetro CRÍTICO para manter a proporção de Churn
)


# ---------------------------------------------------------------------------
# PASSO 2: Verificar o resultado da divisão
# ---------------------------------------------------------------------------
print("--- Divisão em Treino e Teste Concluída ---")
print("\nDimensões dos conjuntos gerados:")
print(f"Features de Treino (X_train): {X_train.shape}")
print(f"Alvo de Treino (y_train):     {y_train.shape}")
print(f"Features de Teste (X_test):   {X_test.shape}")
print(f"Alvo de Teste (y_test):       {y_test.shape}")
print("\n" + "="*80 + "\n")


# ---------------------------------------------------------------------------
# PASSO 3: Confirmar que a estratificação funcionou
# ---------------------------------------------------------------------------
# Vamos calcular a proporção de Churn nos conjuntos de treino e teste.
# Os valores devem ser praticamente idênticos (~26.5%).

print("--- Verificação da Estratificação ---")
print("\nProporção de Churn no conjunto de Treino (y_train):")
print(y_train.value_counts(normalize=True) * 100)

print("\nProporção de Churn no conjunto de Teste (y_test):")
print(y_test.value_counts(normalize=True) * 100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# ---------------------------------------------------------------------------
# MODELO 1: REGRESSÃO LOGÍSTICA
# ---------------------------------------------------------------------------
print("--- Treinando Modelo 1: Regressão Logística ---")

# 1. Cria o objeto do modelo
#    - class_weight='balanced' ajuda o modelo a lidar com o desequilíbrio de classes.
#    - random_state=42 garante a reprodutibilidade.
log_reg = LogisticRegression(class_weight='balanced', random_state=42)

# 2. Treina o modelo usando os dados de treino
log_reg.fit(X_train, y_train)

print("Regressão Logística treinada com sucesso!")
print("\n" + "="*80 + "\n")


# ---------------------------------------------------------------------------
# MODELO 2: RANDOM FOREST CLASSIFIER
# ---------------------------------------------------------------------------
print("--- Treinando Modelo 2: Random Forest ---")

# 1. Cria o objeto do modelo
#    - n_estimators é o número de árvores na floresta. 100 é um bom começo.
#    - class_weight='balanced' também é aplicável aqui.
#    - random_state=42 para reprodutibilidade.
random_forest = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# 2. Treina o modelo usando os mesmos dados de treino
random_forest.fit(X_train, y_train)

print("Random Forest treinado com sucesso!")
```

Modelo 1: Regressão Logística (Modelo Linear)
Este é um modelo clássico e um excelente ponto de partida para problemas de classificação.

Justificativa da Escolha
Ótimo Ponto de Partida (Baseline): É um modelo simples, rápido de treinar e seus resultados são fáceis de interpretar. Ele nos dá uma base de desempenho para comparar com modelos mais complexos.

Interpretabilidade: A Regressão Logística nos permite entender o impacto de cada variável na probabilidade de evasão. É possível extrair coeficientes que dizem, por exemplo, "para cada mês a mais de contrato, a chance de evasão diminui em X%". Isso é muito valioso para o negócio.

Necessidade de Padronização
Sim, este modelo exige dados padronizados. A Regressão Logística funciona calculando uma soma ponderada das variáveis. Se as variáveis tiverem escalas muito diferentes (como Idoso [0, 1] e FaturaTotal [20, 8000]), a variável com a maior escala dominará o cálculo, e o modelo terá dificuldade em encontrar os pesos corretos. A padronização garante que todas as variáveis contribuam de forma justa com base em seu poder preditivo, e não em sua escala.

Modelo 2: Random Forest (Modelo Baseado em Árvore)
Este é um modelo de ensemble, o que significa que ele combina as previsões de várias "árvores de decisão" para chegar a um resultado mais robusto e preciso.

Justificativa da Escolha
Alto Desempenho: O Random Forest é conhecido por sua alta precisão e robustez. Ele geralmente supera modelos lineares em problemas com relações complexas e não-lineares entre as variáveis.

Baixo Risco de Overfitting: Ao combinar muitas árvores, ele evita o problema de "superajuste" (decorar os dados de treino) que uma única árvore de decisão pode ter.

Importância das Variáveis: O modelo consegue nos fornecer um ranking de quais variáveis foram mais importantes para a tomada de decisão, o que também é um insight valioso.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------------------------------------------------------
# PASSO 1: Fazer previsões com os modelos no conjunto de teste
# ---------------------------------------------------------------------------
# O método .predict() usa o modelo treinado para prever o Churn para os dados de teste.
y_pred_log_reg = log_reg.predict(X_test)
y_pred_random_forest = random_forest.predict(X_test)


# ---------------------------------------------------------------------------
# PASSO 2: Avaliar o Modelo de Regressão Logística
# ---------------------------------------------------------------------------
print("--- Avaliação do Modelo 1: Regressão Logística ---")

# Gerando o Relatório de Classificação completo
print("\nRelatório de Classificação (Regressão Logística):")
print(classification_report(y_test, y_pred_log_reg, target_names=['Permaneceu (0)', 'Evadiu (1)']))

# Gerando a Matriz de Confusão
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)
plt.figure(figsize=(7, 5))
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Previsto: Permaneceu', 'Previsto: Evadiu'],
            yticklabels=['Real: Permaneceu', 'Real: Evadiu'])
plt.title('Matriz de Confusão - Regressão Logística', fontsize=14)
plt.ylabel('Valor Real')
plt.xlabel('Valor Previsto')
plt.show()

print("\n" + "="*80 + "\n")


# ---------------------------------------------------------------------------
# PASSO 3: Avaliar o Modelo Random Forest
# ---------------------------------------------------------------------------
print("--- Avaliação do Modelo 2: Random Forest ---")

# Gerando o Relatório de Classificação completo
print("\nRelatório de Classificação (Random Forest):")
print(classification_report(y_test, y_pred_random_forest, target_names=['Permaneceu (0)', 'Evadiu (1)']))

# Gerando a Matriz de Confusão
cm_random_forest = confusion_matrix(y_test, y_pred_random_forest)
plt.figure(figsize=(7, 5))
sns.heatmap(cm_random_forest, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Previsto: Permaneceu', 'Previsto: Evadiu'],
            yticklabels=['Real: Permaneceu', 'Real: Evadiu'])
plt.title('Matriz de Confusão - Random Forest', fontsize=14)
plt.ylabel('Valor Real')
plt.xlabel('Valor Previsto')
plt.show()
```

Análise Crítica e Comparativa dos Modelos Preditivos
Para avaliar a eficácia dos modelos desenvolvidos, utilizei um conjunto de teste separado, correspondente a 30% dos dados, que não foi exposto durante a fase de treinamento. O objetivo foi simular o desempenho dos modelos em dados novos e desconhecidos, permitindo uma avaliação justa e realista de sua capacidade de generalização.

A avaliação foi baseada em um conjunto de métricas de classificação para obter uma visão completa: Acurácia, Precisão, Recall, F1-Score e a Matriz de Confusão. Para um problema de negócio como a evasão, onde identificar corretamente os clientes que de fato cancelam (a classe minoritária) é crucial, foquei especialmente no Recall e no F1-Score como indicadores-chave de desempenho.

Desempenho da Regressão Logística

O modelo de Regressão Logística serviu como uma excelente base de referência (baseline). Graças ao balanceamento de classes aplicado durante o treinamento, ele demonstrou um Recall robusto, indicando uma forte capacidade de identificar uma grande parte dos clientes que realmente cancelaram. No entanto, sua Precisão foi mais moderada, o que significa que, para alcançar essa alta capacidade de detecção, o modelo classificou incorretamente alguns clientes como propensos a cancelar (Falsos Positivos).

Desempenho do Random Forest

O modelo Random Forest apresentou um desempenho superior em praticamente todas as métricas. Ele não apenas alcançou uma Acurácia geral mais alta, mas o mais importante, demonstrou um equilíbrio notavelmente melhor entre Precisão e Recall. Essa harmonia foi refletida em um F1-Score significativamente maior para a classe de evasão, apontando-o como um modelo mais equilibrado e confiável.

Comparação e Conclusão

A análise da Matriz de Confusão de ambos os modelos tornou a superioridade do Random Forest clara. Ele conseguiu reduzir o número de Falsos Negativos – o erro mais crítico para o negócio, que representa clientes que cancelam sem serem previstos – ao mesmo tempo em que manteve um número controlado de Falsos Positivos.

Com base nos resultados, o Random Forest foi selecionado como o modelo de melhor desempenho para prever a evasão de clientes. Sua capacidade de capturar relações mais complexas nos dados, resultando em um F1-Score mais elevado, o torna a ferramenta mais confiável e eficaz para identificar os clientes em risco. Isso permite que a empresa atue de forma mais precisa em suas estratégias de retenção.

Ambos os modelos demonstraram boa capacidade de generalização para os dados de teste, sem sinais evidentes de overfitting significativo. A performance consistente do Random Forest indica que, apesar de sua maior complexidade, ele aprendeu os padrões subjacentes dos dados em vez de simplesmente memorizar o conjunto de treino.

# Análise 1: Relevância das Variáveis na Regressão Logística

A seguir, vamos extrair e visualizar os coeficientes do modelo de Regressão Logística. Esses coeficientes indicam o peso e a direção (positiva ou negativa) da influência de cada variável na probabilidade de um cliente evadir.

```python
# Importando as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# --- Extração e Preparação dos Coeficientes ---
# Os coeficientes estão em log_reg.coef_, que é um array 2D. Pegamos a primeira (e única) linha.
coeficientes = log_reg.coef_[0]
nomes_variaveis = X_train.columns

# Criamos um DataFrame para facilitar a visualização
df_coeficientes = pd.DataFrame({'Variavel': nomes_variaveis, 'Coeficiente': coeficientes})

# Para interpretar a magnitude, usamos o valor absoluto do coeficiente
df_coeficientes['Impacto_Absoluto'] = np.abs(df_coeficientes['Coeficiente'])

# Ordenamos pelo impacto absoluto para ver as variáveis mais influentes
df_coeficientes_sorted = df_coeficientes.sort_values(by='Impacto_Absoluto', ascending=False)

print("--- 10 Variáveis Mais Relevantes (Regressão Logística) ---")
print(df_coeficientes_sorted.head(10))
print("\n" + "="*80 + "\n")


# --- Visualização dos Coeficientes ---
# Vamos plotar as 20 variáveis de maior impacto (as 10 que mais aumentam o risco e as 10 que mais reduzem)

# Separando as top 10 positivas e negativas pelo valor do coeficiente
df_coef_plot = pd.concat([
    df_coeficientes_sorted.sort_values(by='Coeficiente', ascending=False).head(10),
    df_coeficientes_sorted.sort_values(by='Coeficiente', ascending=False).tail(10)
])

plt.figure(figsize=(12, 8))
sns.barplot(x='Coeficiente', y='Variavel', data=df_coef_plot, palette="coolwarm")
plt.title('Impacto das Variáveis na Evasão (Regressão Logística)', fontsize=16, pad=20)
plt.xlabel('Coeficiente (Impacto na Chance de Churn)', fontsize=12)
plt.ylabel('Variáveis', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

### Interpretação dos Resultados - Regressão Logística

A análise dos coeficientes da Regressão Logística nos permite quantificar o impacto de cada característica na evasão. O gráfico de barras acima ilustra os fatores de risco e de proteção mais significativos.

Observa-se que variáveis como `TipoContrato_Month-to-month` possuem um coeficiente positivo elevado, confirmando que a ausência de um contrato de longo prazo é um forte indicador de risco de cancelamento. Por outro lado, variáveis como `MesesContrato` e `TipoContrato_Two year` apresentam coeficientes fortemente negativos, atuando como os principais fatores de retenção de clientes. Isso demonstra que a lealdade e os compromissos de longo prazo são cruciais para a estabilidade da base de clientes.

# Análise 2: Relevância das Variáveis no Random Forest

Diferente da Regressão Logística, o Random Forest não gera coeficientes. Em vez disso, ele calcula a "importância" de cada variável (*feature importance*), que mede o quanto cada uma contribuiu para a redução da impureza e para a precisão geral do modelo em todas as árvores da floresta.

```python
# Importando as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Extração da Importância das Variáveis ---
importancia = random_forest.feature_importances_
nomes_variaveis = X_train.columns

# Criamos um DataFrame para facilitar a visualização
df_importancia = pd.DataFrame({'Variavel': nomes_variaveis, 'Importancia': importancia})

# Ordenamos pela importância para ver as mais relevantes
df_importancia_sorted = df_importancia.sort_values(by='Importancia', ascending=False)

print("--- 15 Variáveis Mais Relevantes (Random Forest) ---")
print(df_importancia_sorted.head(15))
print("\n" + "="*80 + "\n")


# --- Visualização da Importância ---
plt.figure(figsize=(12, 8))
sns.barplot(x='Importancia', y='Variavel', data=df_importancia_sorted.head(15), palette="viridis")
plt.title('Importância das Variáveis na Previsão (Random Forest)', fontsize=16, pad=20)
plt.xlabel('Importância Relativa', fontsize=12)
plt.ylabel('Variáveis', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

### Interpretação dos Resultados - Random Forest

A análise de importância de variáveis do Random Forest confirma e reforça os insights obtidos com a Regressão Logística, porém com uma perspectiva diferente. Em vez de medir a direção do impacto, esta métrica quantifica o poder preditivo geral de cada variável.

Conforme o gráfico, as variáveis mais decisivas para o modelo foram `MesesContrato`, `FaturaTotal` e `FaturaMensal`. Isso indica que o comportamento histórico e financeiro do cliente são os preditores mais fortes da evasão. O tipo de contrato e o método de pagamento também figuram com alta relevância. Essa convergência de resultados entre dois modelos fundamentalmente diferentes nos dá alta confiança sobre quais são os fatores-chave que governam a decisão de um cliente permanecer ou sair.

### 2. Metodologia Aplicada

O processo iniciou-se com a extração e limpeza de dados brutos em formato JSON. O conjunto de dados foi submetido a um extenso pré-processamento, incluindo a expansão de dados aninhados, a codificação de variáveis categóricas (One-Hot Encoding) e a padronização de variáveis numéricas para garantir a compatibilidade com diversos algoritmos.

Foram treinados e avaliados dois modelos:
1.  **Regressão Logística:** Utilizado como um modelo base (*baseline*) por sua simplicidade e interpretabilidade.
2.  **Random Forest:** Um modelo de *ensemble* mais complexo, conhecido por sua alta precisão e capacidade de capturar relações não-lineares.

A avaliação foi realizada em um conjunto de teste (30% dos dados) para garantir uma medição imparcial do desempenho, utilizando métricas como Acurácia, Precisão, Recall e F1-Score.

### 3. Fatores que Mais Influenciam a Evasão

A análise de importância de variáveis de ambos os modelos convergiu, apontando para um conjunto claro de fatores que mais impactam a decisão de um cliente cancelar o serviço.

1.  **Tipo e Duração do Contrato (`MesesContrato`, `TipoContrato`):** Este é, de longe, o fator mais decisivo. Clientes com **contratos mensais (`Month-to-month`)** são massivamente mais propensos a cancelar. Em contrapartida, um maior tempo de contrato (`MesesContrato`) e a adesão a planos de **um ou dois anos** são os mais fortes indicadores de retenção. Isso evidencia que a falta de um compromisso de longo prazo é a principal porta de saída para os clientes.

2.  **Fatura Mensal (`FaturaMensal`):** O valor cobrado mensalmente tem um impacto direto. Clientes com faturas mais altas, especialmente aqueles com serviços de Fibra Óptica (que geralmente têm um custo maior), apresentam uma maior tendência à evasão. Existe uma "zona de perigo" de faturamento onde o risco de cancelamento se torna máximo.

3.  **Método de Pagamento (`MetodoPagamento`):** Clientes que utilizam **boleto eletrônico (`Electronic check`)** como forma de pagamento cancelam com uma frequência significativamente maior do que aqueles que usam métodos de pagamento automático (cartão de crédito e débito em conta). O atrito gerado pela necessidade de uma ação manual de pagamento mensal é um fator de risco.

### 4. Estratégias de Retenção Propostas

Com base nos fatores identificados, as seguintes estratégias de retenção são recomendadas para atuar diretamente nas causas da evasão:

1.  **Ação em Contratos Mensais (Ação de Alto Impacto):**
    * **Estratégia:** Desenvolver campanhas de marketing direcionadas para a base de clientes com contrato `Month-to-month`. Oferecer incentivos claros (ex: um mês de serviço gratuito, um upgrade de plano ou um desconto percentual) para a migração para um contrato de 1 ou 2 anos.
    * **Justificativa:** Ataca diretamente o principal fator de risco, "travando" o cliente em um ciclo de fidelidade mais longo.

2.  **Programa de Onboarding para Novos Clientes:**
    * **Estratégia:** Implementar um programa de engajamento proativo focado nos primeiros 3-6 meses do cliente (o período mais crítico, conforme a análise de `MesesContrato`). O programa deve incluir contatos da equipe de "sucesso do cliente", dicas de uso dos serviços e pesquisas de satisfação para garantir uma percepção de valor positiva desde o início.
    * **Justificativa:** Reduz o churn precoce, que é o mais comum, ao fortalecer o relacionamento inicial com o cliente.

3.  **Incentivo à Automação de Pagamentos:**
    * **Estratégia:** Criar uma campanha oferecendo um pequeno benefício único (ex: R$10 de desconto na próxima fatura) para clientes que migrarem do boleto eletrônico para métodos de pagamento automático.
    * **Justificativa:** Reduz o "atrito" do pagamento, uma causa significativa de cancelamento, e aumenta a conveniência para o cliente, tornando menos provável que ele procure alternativas.

4.  **Ações Preditivas com o Modelo:**
    * **Estratégia:** Integrar o modelo Random Forest a um sistema de CRM. Diariamente, o modelo pode gerar uma pontuação de risco de evasão para cada cliente. Clientes de alto valor com pontuação de risco elevada devem ser automaticamente sinalizados para uma abordagem proativa da equipe de retenção, que pode oferecer uma revisão de plano ou um desconto de lealdade.
    * **Justificativa:** Transforma a análise de reativa para proativa, permitindo que a empresa aja **antes** que o cliente decida cancelar, com base em uma previsão de alta precisão.
